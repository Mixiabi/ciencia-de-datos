```python
import matplotlib.pyplot as plt

# Datos
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

# Crear el gráfico con estilo
plt.plot(x, y, label="Cuadrados", color='green', linestyle='--', marker='o')

# Añadir título y etiquetas
plt.title("Gráfico de Números Cuadrados")
plt.xlabel("Valores de X")
plt.ylabel("Valores de Y")

# Mostrar la leyenda
plt.legend()

# Mostrar el gráfico
plt.show()
```
